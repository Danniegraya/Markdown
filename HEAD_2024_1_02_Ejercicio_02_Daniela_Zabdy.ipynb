```
import csv

años= range(2, 8)
i=2
partidos_bruto= []


for i in años:
    name= "atp_matches_201" +str(i)+ ".csv"
    f= open(name)
    data= csv.reader(f)
    categorias= next(data)
    partidos_bruto+= [p for p in data]


len(partidos_bruto)
```

- Crea una lista con cada partido, donde los datos de cada partido sean un diccionario. Con las categorías como llave y el valor como el contenido.


```
partidos=[]
for p in partidos_bruto:
    partido_d=dict()
    for categoria, dato in zip(categorias, p):
        partido_d[categoria]=dato
    partidos.append(partido_d)

partidos[10000] #posición 0
```

# Ordenamiento
- Obtén los 10 jugadores con más partidos ganados y especifica cuántos partidos ganaron.

```
for partido in partidos:
    ganador = partido['winner_name']

    if ganador in ganados:
        ganados[ganador] += 1
    else:
        ganados[ganador] = 1

tentop_winners = sorted(ganados.items(), key=lambda x: x[1], reverse=True)

tentop_winners = tentop_winners[:10]  # Cambia top_winners a tentop_winners

for jugador, ganados in tentop_winners:
    print(f"{jugador}: {ganados} partidos ganados")
```

- Obtén los 10 jugadores con más partidos perdidos y especifica cuántos partidos perdieron.

```
# top 10
perdedores = {}

for partido in partidos:
    perdedor = partido['loser_name']

    if perdedor in perdedores:
        perdedores[perdedor] += 1
    else:
        perdedores[perdedor] = 1

tentop_losers = sorted(perdedores.items(), key=lambda x: x[1], reverse=True)

tentop_losers = tentop_losers[:10]

for jugador, perdidos in tentop_losers:
    print(f"{jugador}: {perdidos} partidos perdidos")
```
