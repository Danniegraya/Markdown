```python
import os
import pandas as pd

if 'dataset.csv' in os.listdir(os.getcwd()):
    # cargar a un dataframe
    df = pd.read_csv('dataset.csv')
    df.columns = ['index'] + list(df.columns.values)[1::]
else:
    # Si no existe
    df = pd.DataFrame()
    folders = ['20_1', '20_2', '20_3']

    # Iterar
    for folder in folders:
        for file in os.listdir(folder):
            try:
                aux = pd.read_xml(os.path.join(folder, file))
                df = pd.concat([df, aux])
            except:
                continue
    df = df.reset_index()
    df.to_csv('dataset.csv')
print(df.sample(5))
```

```python
# info columnas
print(df.columns)
```

```python
departments = pd.read_xml("departments.xml")
departments
```

```python
# Top 5 salarios con información del departamento:
top_salaries = df.sort_values(by='salary', ascending=False).head(50)
top_salaries = pd.merge(top_salaries, departments, left_on='deptno', right_on='deptno')[['firstname', 'lastname', 'salary', 'deptname']]
top_salaries.head(50)
```

```python
# Cantidad de personas por departamento con nombres
people_per_department = df['deptno'].value_counts().reset_index()
people_per_department.columns = ['deptno', 'PeopleCount']

people_per_department = pd.merge(people_per_department, departments, left_on='deptno', right_on='deptno')[['deptname', 'PeopleCount']]
print(people_per_department)
```

```python
# Empleados más jovenes
youngest_employees = df.sort_values(by='birthdate', ascending=False).head(5)
print(youngest_employees)
```

```python
# Promedio de salario por departamento
average_salary_by_department = df.groupby('deptno')['salary'].mean().reset_index()
average_salary_by_department.columns = ['deptno', 'AverageSalary']
average_salary_by_department = pd.merge(average_salary_by_department, departments, left_on='deptno', right_on='deptno')[['deptname', 'AverageSalary']]

print(average_salary_by_department)
```

```python
# Top 5 salarios
top_5_salaries_by_department = df.groupby('deptno')['salary'].nlargest(5).reset_index()
top_5_salaries_by_department = pd.merge(top_5_salaries_by_department, departments, left_on='deptno', right_on='deptno')[['deptname', 'salary']]

print(top_5_salaries_by_department)
```

```python
# Nombre del departamento y el nombre de su manager
department_managers = pd.merge(df, departments, left_on='empno', right_on='mgrno')[['deptname', 'firstname', 'lastname']]
department_managers = department_managers.set_index('deptname')

print(department_managers)
```

```python
# Calculo de la diferencia entre el salario máximo y mínimo en cada departamento
salary_range_by_department = df.groupby('deptno')['salary'].agg(lambda x: x.max() - x.min()).reset_index()
salary_range_by_department.columns = ['deptno', 'SalaryRange']

salary_range_by_department = pd.merge(salary_range_by_department, departments, left_on='deptno', right_on='deptno')[['deptname', 'SalaryRange']]
print(salary_range_by_department)
```
