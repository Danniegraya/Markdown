```python
corredores = []
file = open('marathon_results_2017.csv') #lo volvemos a abrir para que se vuelva a correr
datos = csv.reader(file)
encabezados = next(datos)

#ahora construire mi diccionario para cada elemento de mis datos
for persona in datos:
    for encabezado, dato in zip(encabezados, persona):
        personaD[encabezado] = dato
    corredores.append(personaD)
```
