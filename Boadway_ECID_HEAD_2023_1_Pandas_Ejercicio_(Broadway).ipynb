```python
#Para la obra de Mary Poppins
# Para cada semana, calcular el valor promedio por boleto
dfMP["Boleto.Promedio"] = dfMP["Statistics.Gross"] / dfMP["Statistics.Attendance"]
dfMP[ ["Date.Full","Boleto.Promedio"] ]
```

```python
#Para la obra de Mary Poppins
#Maximo valor promedio por boleto
dfMP["Boleto.Promedio"] = dfMP["Statistics.Gross"] / dfMP["Statistics.Attendance"]
dfMP[ ["Date.Full","Boleto.Promedio"] ].max()
```

```python
# Para cada teatro
# Calcular la capacidad máxima del teatro

df.groupby("Show.Theatre")["Statistics.Capacity"].max()
```

```python
# Para todos los teatros
# ¿Cuál es el teatro con el nombre más largo?
lista_teatros = df.groupby("Show.Theatre")["Statistics.Capacity"].max().index

#el max es de 22 caracteres, print
print(max( [(len(n), n) for n in lista_teatros] )) #mayores caracteres
print(max( [(len(n.split()), n) for n in lista_teatros] )) #menores caracteres

for nombre in df.groupby("Show.Theatre")["Show.Theatre"].max().index:
    print(nombre, len(nombre))
```
